```python
def outliers(df):
    Q1=df.quantile(0.25)
    Q3=df.quantile(0.75)
    IQR=Q3-Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
   # print(IQR,Q1)
    return (lower_bound,upper_bound)
```

```python
#pd.DataFrame(sdo.groupby(by = ['InvestmentnType'])['NewAmountUSD'].sum()).sort_values(by = 'NewAmountUSD', ascending=False)
#pd.DataFrame(sdo.groupby(by = ['InvestorsName'])['NewAmountUSD'].sum()).sort_values(by = 'NewAmountUSD', ascending=False).head()
#count_start = 
no_of_start = pd.DataFrame(sdo['InvestorsName'].value_counts()).drop(index = ['Undisclosed investors','Undisclosed Investors',
 'nan'],axis =0).reset_index().rename(columns = {'InvestorsName':'No_of_startup_funding',"index":'InvestorsName'}).sort_values(by = 'InvestorsName', ascending=True)
no_of_start.set_index('InvestorsName',inplace =True)
no_of_start.reset_index(inplace = True)

ind = list(no_of_start['InvestorsName'])


amount = pd.DataFrame(sdo.loc[sdo['InvestorsName'].apply(lambda s : True if s in ind else False)].sort_values(by = 'InvestorsName').groupby(by = 'InvestorsName')['NewAmountUSD'].sum())
amount.rename(columns ={'NewAmountUSD': 'sum_of_funding'} , inplace = True)
amount.reset_index(inplace = True)
amount.sort_values(by= 'InvestorsName', inplace = True)
amount

no_of_start['sum'] = amount['sum_of_funding']
print('\n-------------------------Investor funding maximum No. of Startups-------------------------\n',
      no_of_start.sort_values(by = 'No_of_startup_funding', ascending= False).head())
print('\n-------------------------Investor spend maximum amount-------------------------\n',
      no_of_start.sort_values(by = 'sum', ascending= False).head())

```
